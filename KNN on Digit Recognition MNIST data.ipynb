```python

from sklearn.decomposition import PCA
# feature extraction
pca_results=[]
for k in range (1,500,20):
    pca = PCA(n_components=k)
    fit = pca.fit(reduced_train_images)
# summarize componentsii
    pca_train = pca.transform(reduced_train_images)
    neigh = KNeighborsClassifier(n_neighbors=5,n_jobs=-1)
    neigh.fit(pca_train, reduced_train_labels)
    plabels2 = neigh.predict(pca_train)
    pca_results.append([k,accuracy_score(reduced_train_labels, plabels2)])
```

```python
myList = range(1,100,5)
neighbors = filter(lambda x: x % 2 != 0, myList)
from sklearn.model_selection import cross_val_score
```

```python
cv_scores_train = []
cv_scores_test=[]
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k,n_jobs=-1)
    scores_train = cross_val_score(knn, reduced_train_images, reduced_train_labels, scoring='accuracy',n_jobs=-1,verbose=True)
    scores_test = cross_val_score(knn,reduced_test_images,reduced_test_labels,scoring='accuracy',n_jobs=-1,verbose=True)
    cv_scores_test.append(scores_train.mean())
    cv_scores_train.append(scores_test.mean())
```

```python
MSE_test = [1 - x for x in cv_scores_test]
MSE_train=[1 - x for x in cv_scores_train]
# determining best k
optimal_k = neighbors[MSE_test.index(min(MSE_test))]
print "The optimal number of neighbors is %d" % optimal_k

# plot misclassification error vs k
plt.plot(neighbors, MSE_test)
plt.plot(neighbors, MSE_train)
plt.legend(['train error','test eror'], loc='upper left')
plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error')
plt.show()
```
